```python
driver = webdriver.Chrome("./chromedriver")

#kind 접속
driver.get("https://kind.krx.co.kr/corpgeneral/corpList.do?method=loadInitPage#")

# 1초간
driver.implicitly_wait(1)

# 항목 마지막 번호 
x = driver.find_element_by_xpath('//*[@id="main-contents"]/section[2]/div[1]')
divTag = driver.find_elements_by_css_selector("div em")
maxSize = int(divTag[0].text.replace(",","")) #2,499개
name = []
code = []
ran = round(maxSize/15)
for a in range(ran+1):
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")
    
    for a in soup.find_all('td', class_='first'):
        print(a.a.text)
        #기업이름에 공백이 들어가 있기 때문에 trim
        name.append(a.a.text.strip())
        att = a.select('a')
        for z in att:
            #print(z['onclick']) #companysummary_open('36232'); return false;
            print(z['onclick'].replace("companysummary_open('","").replace("'); return false;",""))
            code.append(z['onclick'].replace("companysummary_open('","").replace("'); return false;",""))
        
    # 3초간
    driver.implicitly_wait(3)
    time.sleep(2)
    TagA = driver.find_elements_by_css_selector('article section div a')
    
    TagA[-3].click()
```

```python
len(name)
len(code)
```

```python
# kind parse data
p2 = pd.DataFrame({"corp_name":name, "kind_code":code})
p2["kind_code"]
```

```python
result = pd.merge(p1,p2, on="corp_name")
result["stock_code"]=result["stock_code"].astype(str)
result["kind_code"]=result["kind_code"].astype(str)
result
```
