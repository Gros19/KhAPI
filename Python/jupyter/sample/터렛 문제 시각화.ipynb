```python
import matplotlib.pyplot as pl
import math


#찌그러지는 문제...
def draw_circle_graphs(x1, y1, r1, x2, y2, r2):
    xlist1 = []
    xlist2 = []
    ylist1 = []
    ylist2 = []
    for theta in range(0, 360):
        xlist1.append(x1+r1*math.cos(math.radians(theta)))
        ylist1.append(y1+r1+math.sin(math.radians(theta)))
        xlist2.append(x2+r2*math.cos(math.radians(theta)))
        ylist2.append(y2+r2+math.sin(math.radians(theta)))
        
    pl.figure(figsize=(50, 50))
    pl.plot(xlist1,ylist1,xlist2,ylist2)
    pl.grid()
    pl.xlabel('x-axis')
    pl.ylabel('y-axis')
    title = 'circle equation graph'
    pl.title(title)
    pl.savefig('circle.png')
```

```python
draw_circle_graphs(-1, 8, 10, -1, 0, 10)
```

```python
def draw_circle_graph(a, b, r): 
    '''
    원의 방정식 그래프 그리기 
      - 원의 중심 C(a, b), 원의 반지름 r을 함수 인자로 입력 
    '''

    x = []   # x 좌표값을 담을 list 정의
    y = []   # y 좌표값을 담을 list 정의
      
    for theta in range(0, 360):
        # 0도에서 360도까지 1도(degree) 간격으로,
        x.append(a + r * math.cos(math.radians(theta)))      # math 모듈 사용 
        y.append(b + r * math.sin(math.radians(theta)))       # math 모듈 사용
    
    pl.figure(figsize=(10, 10))   # x, y 축의 비율을 동일하게 설정 
    pl.plot(x, y)   # x, y 값을 담은 그래프 object 만들기 
    pl.grid()
    pl.xlabel('x-axis')
    pl.ylabel('y-axis')

    # title 출력
    title = 'circle equation graph'      
    pl.title(title)       
```

```python
draw_circle_graph(-4,5, 10)
```

```python
import numpy as np
import matplotlib.patches as patches
import matplotlib.pyplot as plt
def Draw_circle(x1, y1, r1, x2, y2, r2):
    X = np.array([-5, 5])
    Y = np.array([-5, 5])
    plt.plot(X,Y,color="None")

    #circle
    shp = patches.Circle((x1, y1), radius=r1, color='b')
    plt.gca().add_patch(shp)
    
     #circle
    shp = patches.Circle((x2, y2), radius=r2, color='g')
    plt.gca().add_patch(shp)
    
    plt.axis('scaled')
    plt.show()
    
```

```python
# -1 8 10 -1 0 10
# -6 3 9 -1 7 6
# -2 3 8 1 2 5
# 2 9 6 10 9 10
Draw_circle(-1, 8, 10, -1, 0, 10)
```

```python
Draw_circle(-6, 3, 9, -1, 7, 6)
```

```python
Draw_circle(-2, 3, 8, 1, 2, 5)
```

```python
Draw_circle(2, 9, 6, 10, 9, 10)
```
